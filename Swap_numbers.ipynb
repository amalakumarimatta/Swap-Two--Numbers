## **Swap numbers in multiple ways using python language**

## **Method 1:** Using Temporary Variable

```python
a = 10
b = 20

temp = a
a = b
b = temp

print("a =", a)
print("b =", b)
```

## **Method 2:** Without Temporary Variable

```python
a = 10
b = 20

a, b = b, a

print("a =", a)
print("b =", b)
```

## **Method 3:** Using Addition and Subtraction

```python
a = 10
b = 20

a = a + b
b = a - b
a = a - b

print("a =", a)
print("b =", b)
```

## **Method 4:** Using Multiplication and Division

```python
a = 10
b = 20

a = a * b
b = a / b
a = a / b

print("a =", int(a))
print("b =", int(b))
```

## **Method 5:** Using XOR Operator

```python
a = 10
b = 20

a = a ^ b
b = a ^ b
a = a ^ b

print("a =", a)
print("b =", b)
```

## **Method 6:** Using Function

```python
def swap(x, y):
    return y, x

a = 10
b = 20

a, b = swap(a, b)

print("a =", a)
print("b =", b)
```



## **Method 7:** Using Lambda Function

```python
swap = lambda x, y: (y, x)

a = 10
b = 20

a, b = swap(a, b)

print("a =", a)
print("b =", b)
```
